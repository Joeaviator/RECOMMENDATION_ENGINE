```python
import pandas as pd
```

```python
#Load datasets
movies_ds=pd.read_csv("movies.csv")
rating_ds=pd.read_csv("ratings.csv")

```

```python
#Read the datasets
movies_ds.head()
```

```python
rating_ds.head()
```

```python
movies_ds['title'] = movies_ds['title'].str.replace(r'\s*\(\d{4}\)', '', regex=True)
movies_ds['title'] = movies_ds['title'].str.replace(r'\s*\(\)\s*', '', regex=True).str.strip()
```

```python
movies_ds.head()
```

```python
#remove | this from genres
movies_ds['genres']=movies_ds['genres'].str.split('|')
movies_ds.head()
```

```python
rating_ds.drop('timestamp',axis=1,inplace=True)
rating_ds.head()
```

```python
#One hot encoding-convert categories into numbers
movies_copy=movies_ds.copy()
for i,row in movies_ds.iterrows():
  for genre in row['genres']:
    movies_copy.at[i,genre]=1
#replace the null values in movies copy
movies_copy=movies_copy.fillna(0)
```
